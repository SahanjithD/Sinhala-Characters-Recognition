```python
import tkinter as tk
from PIL import ImageTk,Image,ImageDraw
import numpy as np
import cv2
import os
import joblib

model=joblib.load('sinhala_char_model.sav')

win = tk.Tk()

width =500
height =500
font_btn ='Arial  10 bold'
font_label ='Arial  12 bold'
count =0
labelDic = {0:'අ', 1:'ආ' ,2:'ඇ', 3:'ඉ'}


def event_function(event):
    x=event.x
    y=event.y
    
    x1=x-30
    x2=x+30
    
    y1=y-30
    y2=y+30
    canvas.create_oval((x1,y1,x2,y2),fill='black')
    img_draw.ellipse((x1,y1,x2,y2),fill='black')


def save():
    global count
    img_array= np.array(img)
    path = os.path.join('data', str(count) + '.jpg')
    # img_array=cv2.resize(img_array,(8,8))
    cv2.imwrite(path,img_array)
    count +=1
    
    
def clear():
    global img,img_draw
    canvas.delete('all')
    img=Image.new('RGB', (width,height) ,(255,255,255))
    img_draw=ImageDraw.Draw(img)
    label_status.config(text= 'PREDICTED CHARACTER: NONE')
    

def predict():
    img_array=np.array(img)
    img_array= cv2.cvtColor(img_array,cv2.COLOR_BGR2GRAY)
    img_array= cv2.resize(img_array,(8,8))
    img_array= np.reshape(img_array,(1,64))
    
    result=model.predict(img_array)[0]
    print(result)
    label=labelDic[result]
    label_status.config(text= 'PREDICTED CHARACTER: ' +label)


canvas = tk.Canvas(win,width = width,height = height , bg ='white')
canvas.grid(row =0,column =0, columnspan=4)

button_save = tk.Button(win,text ='SAVE', bg='black', fg= 'white', font= font_btn ,command = save)
button_save.grid(row=1,column =0)

button_predict = tk.Button(win,text ='PREDICT', bg='black', fg= 'white', font= font_btn ,command=predict)
button_predict.grid(row=1,column =1)

button_clear = tk.Button(win,text ='CLEAR', bg='black', fg= 'white', font= font_btn , command= clear)
button_clear.grid(row=1,column =2)

button_exit = tk.Button(win,text ='EXIT', bg='black', fg= 'white', font= font_btn ,command=win.destroy)
button_exit.grid(row=1,column =3)

label_status =tk.Label(win,text= 'PREDICTED CHARACTER: NONE', bg ='white', font= font_label)
label_status.grid(row=3, column = 0, columnspan =4)

canvas.bind('<B1-Motion>',event_function)

img=Image.new('RGB', (width,height) ,(255,255,255))
img_draw =ImageDraw.Draw(img)

win.mainloop()
```
